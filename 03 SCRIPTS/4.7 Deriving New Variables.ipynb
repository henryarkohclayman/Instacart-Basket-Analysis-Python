1. Creating "price_label" column

```python
# Create a subset of first million rows
df = df_ords_prods_merge[:1000000]
```

```python
# Define a function for "price label"
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# return value counts of price_range
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# return value counts of price_range
df['price_range'].value_counts(dropna = False)
```

```python
# return max value of prices
df['prices'].max()
```

If-Statements with the loc() Function

```python
# Use loc() method on df subset to apply price range function.
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# return value counts of price_range_loc
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Use loc() method on entire ords_prods_merge dataframe to apply price range function.(High range)
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Creating mid range label on full ords_prods_merge

df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Creating low range label on full ords_prods_merge

df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

# 4. Create new column "busiest_period_of_day" to identify time periods “Most orders,” “Average orders,” and “Fewest orders.”

```python
# Check value counts in "order_hour_of_day" column.
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

The value counts of order_hour_of_day have been categorized into three groups based on order volume. The hours with the highest frequencies—specifically 10, 11, 14, 15, 13, 12, 16, and 9—were labeled as “Most orders.” The hours with the lowest frequencies—23, 6, 0, 1, 5, 2, 4, and 3—were labeled as “Fewest orders.” All remaining hours were classified under the label “Average orders.

```python
# Create for-loop if statement labeling periods of time as “Most orders,” “Average orders,” and “Fewest orders.”
result_3 = []
for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]: result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]: result_3.append("Fewest orders")
  else: result_3.append("Average orders")
```
